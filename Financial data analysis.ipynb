```python
from numpy.random import default_rng

rg = default_rng(12345)

X = rg.uniform(low=0.0, high=10.0, size=(1000,))

lr[0:10]
```

```python
slope = 0.25
intercept = -1.25

y = slope * X + rg.normal(loc=0.0, scale=1.0, size=(1000,)) + intercept
```

```python
plt.scatter(X, y, s=1)

plt.xlabel('X')
plt.ylabel('Y')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(fit_intercept=True, normalize=False, copy_X=True, n_jobs=None)
```

```python
lr.fit(X.reshape(-1,1), y)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X.reshape(-1,1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                    f_data['EDUCATION'].values.reshape(-1,1),
                                    f_data['default payment next month'].values,
                                    test_size = 0.20,
                                    random_state = 24
                        )
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np

lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
is_correct = y_pred == y_test
```

```python
np.round(np.mean(is_correct), 3)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```
